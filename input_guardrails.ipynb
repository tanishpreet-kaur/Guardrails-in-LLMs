```python
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
banned_words = ["hack", "password", "malware"]

def input_guardrail(query):
    for word in banned_words:
        if word in query.lower():
            return "Blocked unsafe query"
    return query
```

```python
llm = init_chat_model("google_genai:gemini-2.5-flash-lite")
```

```python
query = "How can i hack into my friend's computer?"
```

```python
def guarded_invoke(query):
    safe_query = input_guardrail(query)
    if not safe_query:
        return "Unsafe input detected."
    return llm.invoke(query)
```

```python
ans = guarded_invoke(query)
```
