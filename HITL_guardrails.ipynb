```python
from langchain.chat_models import init_chat_model
from langchain_core.messages import HumanMessage
```

```python
llm = init_chat_model("google_genai:gemini-2.5-flash-lite")
```

```python
HIGH_RISK_ACTIONS = [
    "delete",
    "transfer money",
    "send email",
    "shutdown",
    "drop database"
]
```

```python
def requires_human_approval(text):
    text = text.lower()

    for action in HIGH_RISK_ACTIONS:
        if action in text:
            return True

    return False
```

```python
def human_approval(query):
    print("\n⚠️ HIGH RISK ACTION DETECTED")
    print(f"Query: {query}")
    approval = input(
        "\nApprove this action? (yes/no): "
    )
    return approval.lower() == "yes"
```

```python
def safe_chat(user_input):
    if requires_human_approval(user_input):
        approved = human_approval(user_input)

        if not approved:
            return "❌ Action rejected by human."

    response = llm.invoke([
        HumanMessage(content=user_input)
    ])
    
    return response.content
```

```python
query = "Send email to finance team"
result = safe_chat(query)
print("\nAssistant:")
print(result)
```
